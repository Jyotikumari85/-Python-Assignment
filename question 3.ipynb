```python
# simple rock- paper-scissors game
 #play_round function to determine the winner
def play_round(player1, player2):
    if player1 == player2:
        return "Draw"
    elif (player1 == "R" and player2 == "S") or (player1 == "S" and player2 == "P") or (player1 == "P" and player2 == "R"):
        return "Player1 wins"
    else:
        return "Player2 wins"

def play_game():
    player1_wins = 0
    player2_wins = 0

#players can choose their moves using input
#set up the game to have 10 rounds
    for round in range(1, 11):
        print(f"Round {round}:")
        player1 = input("Player 1, enter your choice (R/P/S): ").upper()
        player2 = input("Player 2, enter your choice (R/P/S): ").upper()

#Adding some prompts to indicate who won each round or the game overall
        result = play_round(player1, player2)
        print(f"Result: {result}")

        if result == "Player1 wins":
            player1_wins += 1
        elif result == "Player2 wins":
            player2_wins += 1

        print()

    print("Final Results:")
    print(f"Player 1 wins: {player1_wins}")
    print(f"Player 2 wins: {player2_wins}")

    if player1_wins > player2_wins:
        print("Player 1 is the winner!")
    elif player2_wins > player1_wins:
        print("Player 2 is the winner!")
    else:
        print("It's a draw!")

play_game()
```
